```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import math

def get_user_input():
    mean = float(input("Enter the mean: "))
    sd_expr = input("Enter the stdev: ")
    decimal = int(input("Enter the decimal place (e.g., 4 for 4 decimal places): "))

    try:
        sd = eval(sd_expr, {"__builtins__": None}, {"sqrt": math.sqrt})
    except Exception as e:
        print(f"Invalid standard deviation input: {e}")
        return None, None, None

    return mean, sd, decimal

def normal_distribution_calculator():
    mean, sd, decimal = get_user_input()
    if mean is None or sd is None:
        return

    print("Choose calculation option:")
    print("1. Probability to the left of x")
    print("2. Probability to the right of x")
    print("3. Probability between two values")

    option = int(input("Enter your choice (1/2/3): "))

    def probability_left(x, mean, sd):
        return round(stats.norm.cdf(x, mean, sd), decimal)

    def probability_right(x, mean, sd):
        return round(1 - stats.norm.cdf(x, mean, sd), decimal)

    def probability_between(a, b, mean, sd):
        return round(stats.norm.cdf(b, mean, sd) - stats.norm.cdf(a, mean, sd), decimal)

    # Generate x values for the normal distribution
    x = np.linspace(mean - 4 * sd, mean + 4 * sd, 1000)
    y = stats.norm.pdf(x, mean, sd)
    plt.plot(x, y, label="Normal Distribution", color='blue')

    if option == 1:
        x_val = float(input("Enter the value of x: "))
        result = probability_left(x_val, mean, sd)
        print(f"P(X < {x_val}) = {result}")
        plt.fill_between(x, y, where=(x <= x_val), color='lightblue')
        plt.axvline(x_val, color='red', linestyle='--')

    elif option == 2:
        x_val = float(input("Enter the value of x: "))
        result = probability_right(x_val, mean, sd)
        print(f"P(X > {x_val}) = {result}")
        plt.fill_between(x, y, where=(x >= x_val), color='lightblue')
        plt.axvline(x_val, color='red', linestyle='--')

    elif option == 3:
        a = float(input("Enter the lower bound (a): "))
        b = float(input("Enter the upper bound (b): "))
        result = probability_between(a, b, mean, sd)
        print(f"P({a} < X < {b}) = {result}")
        plt.fill_between(x, y, where=(x >= a) & (x <= b), color='lightblue')
        plt.axvline(a, color='red', linestyle='--')
        plt.axvline(b, color='red', linestyle='--')

    else:
        print("Invalid choice. Please select 1, 2, or 3.")

    # Print professor info at the bottom
    print("\nProfessor: Edward Pineda-Castro")
    print("Department of Mathematics")
    print("Los Angeles City College")

    plt.xlabel("X")
    plt.ylabel("Density")
    plt.title("Normal Distribution")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    normal_distribution_calculator()
```
